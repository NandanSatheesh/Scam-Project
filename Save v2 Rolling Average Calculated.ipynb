```python
window_size = 10 
window = np.ones(int(window_size))/window_size
window
```

```python
moving_average = np.convolve(y , window , 'same')
print(moving_average)
```

```python
average = moving_average.tolist()
average[:10]
```

```python
residual = y - average 
residual[:10]
```

```python
std = np.std(residual)
std
```

```python
sigma = 1 
testing_std = residual.rolling(window_size).std()
testing_std[:10]
```

```python
testing_std_df = pd.DataFrame(testing_std)
testing_std_df.head(20)
```

```python
testing_std_df = testing_std_df.dropna(axis=0,how='all')
testing_std_df.head(10)
```
